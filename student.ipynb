```python
import pandas as pd 
import numpy as ny
import matplotlib as plt
%matplotlib inline
```

```python
df = pd.read_csv("movies.csv")
df.head()
```

```python
#finding out if the whole document has been imported correctly.
df.tail()
```

```python
#checking how the document has been built by running the shape method
df.shape
```

```python
# Now we want find more information about our dataset
df.info()
```

```python
#we want to find out null values and missing values on our dataset.
df.isnull().sum()
```

```python
#since the rating column is not needed for this analysis and the high number 
# we drop the rating column and other unnecessary columns 
columns_to_drop = ["votes","writer","rating"]
df = df.drop(columns_to_drop, axis=1)

```

```python
# we look if the changes took place
df.head()
```

```python
#let us find out if the shape changed
df.shape
```

```python
# we would to find out the company that produces the highest number of movies
df["company"].value_counts()
```

```python
# now we wold like to find ot the most expensive movies
#let us assume 500 million is expensive enough
expensive_movies = (df["gross"] > 500000000)
df.loc[expensive_movies]
```

```python
#now we woud like to see only the name ,genre,score and company tha produce these expensive movies
df.loc[expensive_movies,["name", "genre", "score", "company"]]
```

```python
#now we would like to find out if the more expensive the movie is the higher rating

# Create a subset of the DataFrame with only the selected columns and rows
subset_df = df.loc[expensive_movies, ["gross", "score"]]

# Plot the data as a bar graph
subset_df.plot(kind='barh', x='score', y='gross', figsize=(20, 35))

plt.xlabel('score')
plt.ylabel('gross')
plt.title('Gross Expenditure vs. Movie Score for Expensive Movies')

plt.show()
```
